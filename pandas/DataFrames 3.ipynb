### The Review of .apply() method on single column

```python
bond = pd.read_csv("input/jamesbond.csv")
bond.set_index("Film",inplace = True)
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def convert_to_string_add_million(number):
    return str(number) + " MILLIONS!"
bond["Box Office"].apply(convert_to_string_add_million)
```

```python
columns = ["Box Office","Budget","Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_add_million)
```

```python
bond.head(3)
```

### The .apply() method with row values

```python
bond = pd.read_csv("input/jamesbond.csv")
bond.set_index("Film",inplace = True)
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(row):
    budget = row[4]
    collection = row[3]
    if(budget < collection/10):
        return "Best movie"
    elif (budget < collection/4):
        return "Enjoyable"
    else:
        "I have no clue"

bond.apply(good_movie,axis = "columns")    
```
