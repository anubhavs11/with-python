### Array Processing

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
points = np.arange(-5,5,0.01)
points[:5]
```

```python
dx ,dy = np.meshgrid(points,points)
```

```python
dx
```

```python
dy
```

```python
z = (np.sin(dx) + np.sin(dy))
z
```

```python
plt.imshow(z)
```

```python
plt.imshow(z)

plt.colorbar()
plt.title("Plot for sin(x)+sin(y)")
```

```python
A = np.array([1,2,3,4])
B = np.array([100,200,300,400])
condition = np.array([True,True,False,False])
```

```python
ans = np.array([A_value if condition else B_value for A_value,B_value,condition in zip(A,B,condition) ])
ans
```

```python
ans2 = np.where(condition,A,B) # when condition is True choose A otherwise choose B
ans2
```

### Array Input and Output

```python
arr = np.arange(5)
arr
```

```python
np.save("myarray",arr)
```

```python
arr = np.arange(10)
arr
```

```python
np.load("myarray.npy") #loading array from the save file with 'npy' extension numpy file
```

```python
arr
```

```python
arr2 = np.arange(6)
arr2
```

```python
np.savez("ziparray.npz" , x = arr,y = arr2) #saving both arrys in a zip (numpy zip -> npz)
```

```python
archive_array = np.load("ziparray.npz")
```

```python
archive_array["x"]
```

```python
archive_array["y"]
```
