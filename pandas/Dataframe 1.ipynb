### Drop rows with Null values

```python
nba = pd.read_csv("input/nba.csv")
nba.head(3)
```

```python
nba.dropna()
# it will delete all rows having Null value in any column
```

```python
nba.dropna(how = "all",inplace = True)  #by default how = 'any'
# it will delete all rows having Null value in all column
```

```python
nba.dropna(inplace = False) #by deafult inplace = True
```

```python
nba.dropna(axis = "columns")
nba.dropna(axis = 1) #same result
# by dafault axis = 0(rows)
# it will delete any column having null value at any time
```

```python
nba.dropna(subset = ["Salary"])
# delete rows having null value in salary column
```

```python
nba.dropna(subset = ["Salary","College"])
# delete rows having null value in salary column or College column
```

```python
nba["Salary"] = nba["Salary"].astype("int")
```

```python
nba.Salary.head(3)
```

```python
nba["Weight"] = nba["Weight"].astype("int")
nba["Age"] = nba["Age"].astype("int")
nba["Number"] = nba["Number"].astype("int")
```

```python
nba.info()
```

### Sort a Dataframe with .sort_values() method

```python
nba = pd.read_csv("input/nba.csv")
nba.head(3)
```

```python
nba.sort_values("Name").head(4)
```

```python
nba.sort_values(by = "Name",ascending = False).head(4)
```

```python
nba.sort_values("Age").head(4)
```

```python
nba.sort_values("Salary",ascending = True).tail(4)   # by default na_position = 'last'
# Null values will be sorted at the end
```

```python
nba.sort_values("Salary",ascending = False).tail(4) # by default na_position = 'last'
```

```python
nba.sort_values("Salary",na_position = "first").head(2)
```

```python
nba.sort_values("Salary",na_position = "first").tail(2)
```

```python
nba = pd.read_csv("input/nba.csv")
nba.head(3)
```

```python
nba.sort_values(["Team","Name"]).head(3)
# first sorting by Team then , foreach Team sorting by Name is done in ascending order
```

```python
nba.sort_values(["Team","Name"],ascending = False).head(3)
# first sorting by Team then , foreach Team sorting by Name is done in descending order
```

```python
nba.sort_values(["Team","Name"], ascending = [True,False] ).head(3)
# first sorting by Team in ascending then , foreach Team sorting by Name is done in descending order
```

```python
nba.sort_values(["Team","Name"], ascending = [False,True] ).head(3)
# first sorting by Team in descending then , foreach Team sorting by Name is done in ascending order
```

### Rank Values with .rank() method

```python
nba = pd.read_csv("input/nba.csv")
nba = nba.dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
nba["Salary"].rank().head(3) 
# by default ranking is done in ascending order
```

```python
nba["Salary"].rank(ascending = False).head(3)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
```

```python
nba.head()
```

```python
nba.sort_values("Salary",ascending = False).head()
```
